# Preparing data
split into train, validate and test set

```python
#Split data into fraud, non_fraud, and total
#splitting X and Y
X = df.iloc[:, :-1]
Y = df['Class']
#splitting fraud and non-fraud X for future use if necessary
X['Class'] = df['Class']
X_nonFraud = X.loc[df['Class'] == 0]
X_Fraud = X.loc[df['Class'] == 1]
X_nonFraud = X_nonFraud.iloc[:, :-1]
X_Fraud = X_Fraud.iloc[:, :-1]
#splitting fraud and non-fraud Y for future if necessary
Y_Fraud = Y[Y == 1]
Y_nonFraud = Y[Y == 0]
X = df.iloc[:, :-1]
print("Number of non fraud: {}".format(X_nonFraud.shape[0]))
print("Number of fraud: {}".format(X_Fraud.shape[0]))
```

```python
#splitting into x_train, x_val and x_test
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
scaler = preprocessing.MinMaxScaler()
scaled_data = scaler.fit_transform(X)
x_train, x_test, y_train, y_test = train_test_split(scaled_data,Y,test_size=0.2,random_state = 40, stratify = Y)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train,test_size=0.2,random_state = 40, stratify = y_train)
```

# Autoencoder method for anomaly detection
unsupervised method

```python
from tensorflow.keras import models, layers, activations, optimizers, metrics, losses
import numpy as np
import seaborn as sns
import matplotlib.lines as lines
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
#constructing autoencoder model
n_features = X.shape[1]
encoder = models.Sequential(name='encoder')
encoder.add(layer=layers.Dense(units=10, activation=activations.relu, input_shape=[n_features]))
encoder.add(layer=layers.Dense(units=5, activation=activations.relu))
encoder.add(layer=layers.Dense(units=2, activation=activations.relu))

decoder = models.Sequential(name='decoder')
decoder.add(layer=layers.Dense(units=5, activation=activations.relu, input_shape=[2]))
decoder.add(layer=layers.Dense(units=10, activation=activations.relu))
decoder.add(layer=layers.Dense(units=n_features, activation=activations.sigmoid))

autoencoder = models.Sequential([encoder, decoder])
autoencoder.compile(
    loss=losses.MSE,
    optimizer=optimizers.Adam(),
    metrics=[metrics.mean_squared_error])
```

```python
#training autoencoder
history = autoencoder.fit(x=x_train, y=x_train, epochs=10, validation_data=(x_val, x_val))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

# Selection of threshold level

## Selection of threshold if unsupervised
if there is no x_test and y_test

```python
#determining threshold level
plot_samples = 10000
true_x = x_train[:plot_samples].reshape(plot_samples, n_features)
pred_x = autoencoder.predict(x=true_x)
mse = losses.mean_squared_error(true_x, pred_x)
mse_df = pd.DataFrame({
    'mse': mse,
    'n': np.arange(0, plot_samples)})
plot = sns.scatterplot(x=mse_df.n, y=mse_df.mse)

line = lines.Line2D(
    xdata=np.arange(0, plot_samples),
    ydata=np.full(plot_samples, 0.02),
    color='#CC2B5E',
    linewidth=1.5,
    linestyle='dashed')

plot.add_artist(line)
plt.show()
```

```python
#predict function for classification
def predict(x, threshold):
    real_x = x.reshape(x.shape[0], x.shape[1])
    predicted_x = autoencoder.predict(x=real_x)
    mse = losses.mean_squared_error(real_x, predicted_x)
    y_predict = []
    for predict in mse:
        if predict <= threshold:
            y_predict.append(0)
        else:
            y_predict.append(1)
    return np.asarray(y_predict)
```

```python
y_pred = predict(x_test,0.015)
print("Preliminary Results for Autoencoder model")
print(classification_report(y_test, y_pred))
print("AUROC score:{}".format(roc_auc_score(y_test, y_pred)))
print("F1 score:{}".format(f1_score(y_test, y_pred)))
```

## Selecting threshold if supervised
if presence of x_test and y_test
for validation

```python
def findBestThreshold(x_test,y_test,range_):
    auroc = []
    f1 = []
    threshold = []
    increment = (range_[1] - range_[0])/range_[2]
    for i in range(range_[2]):
        current_thresh = range_[0] + increment*(i+1)
        threshold.append(current_thresh)
        y_pred = predict(x_test,current_thresh)
        auroc.append(roc_auc_score(y_test,y_pred))
        f1.append(f1_score(y_test,y_pred))
    threshold_df = pd.DataFrame({
        'threshold': threshold,
        "auroc": auroc,
        "f1": f1
    })
    print(threshold_df)
    sns.lineplot(x='threshold', y='value', hue='variable', 
             data=pd.melt(threshold_df, ['threshold']))
    #sns.lineplot(x = "threshold", y = metric_, data = threshold_df)
        
```

```python
findBestThreshold(x_test,y_test, [0.01,0.03,20])
```
